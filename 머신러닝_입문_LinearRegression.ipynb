# 데이터프레임 만들기

```python
x = [-3,  31,  -11,  4,  0,  22, -2, -5, -25, -14]
y = [ -2,   32,   -10,   5,  1,   23,  -1,  -4, -24,  -13]
print(x)
print(y)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.show()
```

```python
df = pd.DataFrame({'X':x, 'Y':y})
df.shape
```

## 머신러닝 사이킷런 LinearRegression 모형

```python
train_features = ['X']
target_cols = ['Y']
##인덱스 추가
X_train = df.loc[:, train_features]
y_train = df.loc[:, target_cols]
print(X_train)
print(y_train)
print(X_train.shape, y_train.shape)
```

```python
plt.scatter(X_train, y_train)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
##학습된 계수의 의미 _ 사용
lr.coef_, lr.intercept_
```

```python
##array([[1.]]): 2차원 배열이며, 실수(float)로 반환, array([1.])):1차원배열이며 실수로 반환
print ("기울기: ", lr.coef_[0][0])
print ("y절편: ", lr.intercept_[0])
```

```python
import numpy as np
##predict()은 2차원 배열을 요구, 숫자 11을 2차원 배열로 변환
X_new = np.array(11).reshape(1, 1)## (행 열)
lr.predict(X_new)
```

```python
##오차계산하기 (오차 제곱합으로 계산)
mean_error = np.mean((y_train-lr.predict(X_train))**2)
print(mean_error)
```

```python
##오차계산하기
from sklearn.metrics import mean_squared_error
mean_squared_error(y_train, lr.predict(X_train))
```

```python
## 데이터 분포 위에 예측선 그리기
# 예측선용 X 범위 생성
X_line = np.linspace(X_train.min(), X_train.max(), 100).reshape(-1, 1)
y_pred = lr.predict(X_line)

# 그래프 그리기
plt.scatter(X_train, y_train, color='blue', label='datas')  # 산점도
plt.plot(X_line, y_pred, color='red', label='predicts-line')  # 예측선
plt.legend()
plt.title(' Datas & Prediction line')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
